## –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö

### –ü–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

–ü—Ä–∏–≤–µ–¥–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∫ "–∑–º–µ–∏–Ω–Ω–æ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É"

```python
def lower_snake(df):
    df.columns = [x.lower().replace(' ', '_') for x in df.columns]
    print("–°–ø–∏—Å–æ–∫ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:")
    print(df.columns, '\n')

# –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è market_df
lower_snake(market_df)


# –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è market_money_df
lower_snake(market_money_df)


# –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è market_time_df
lower_snake(market_time_df)


# –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –Ω–∞–∑–≤–∞–Ω–∏–π –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è money_df
lower_snake(money_df)
```

### –û–±—Ä–∞–±–æ—Ç–∫–∞ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤

```python
def find_duplicates(df):
    print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ = {df.duplicated().sum()}")

# –ü–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ "market_df"
find_duplicates(market_df) 

# –ü–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ "market_money_df"
find_duplicates(market_money_df)

# –ü–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ "market_time_df"
find_duplicates(market_time_df)

# –ü–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥—É–±–ª–∏–∫–∞—Ç–æ–≤ –≤ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–µ "money_df"
find_duplicates(money_df)
```

### –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤

```python
def values_in_cat_feat(df):
    # –ø–æ–ª—É—á–∞–µ–º —Å–ø–∏—Å–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    cat_col_names = df.select_dtypes(exclude='number').columns.tolist()
    if len(cat_col_names) == 0:
        print("–ù–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")
    
    for column in cat_col_names:
        # –≤—ã–≤–µ–¥–∏—Ç–µ –Ω–∞–∏–º–µ–Ω–æ–≤–∞–Ω–∏–µ —Å—Ç–æ–ª–±—Ü–∞
        print('–ü—Ä–∏–∑–Ω–∞–∫ -', column)
        
        # –ø—Ä–∏–≤–µ–¥–µ–º –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É –≤—Å–µ –∑–Ω–∞—á–µ–Ω–∏—è
        df[column] = df[column].str.lower()
        
        # —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
        print(df[column].value_counts(), '\n')
```

```python
def visualize_categorical_feature(df, feature = None):
    """
    –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ—Ç –∏ –∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏.
    
    :param df: DataFrame —Å –¥–∞–Ω–Ω—ã–º–∏
    :param feature: –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
    """
    counts = df[feature].value_counts()

    # –ì—Ä–∞—Ñ–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
    plt.figure(figsize=(6, 6))
    plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140);
    plt.title(f'{feature}')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()

    # –°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞
    print(f'–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞ –¥–ª—è {feature}:')
    print(f'- –£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è: {df[feature].nunique()}')
    print()
    print(f'- –ú–æ–¥–∞: {df[feature].mode()[0]}')
    print()
    print(f'- –ß–∞—Å—Ç–æ—Ç–∞ –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:\n{df[feature].value_counts()}')
    print()
    print(f'- –ü—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ:\n{df[feature].value_counts(normalize=True) * 100}\n')
```

```python
def visualize_numerical_feature(df, feature = None, xlabel = None, title = None, bins = 30, hue=None):
    """
    –°–æ–∑–¥–∞–µ—Ç –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫, —Å–æ—Å—Ç–æ—è—â–∏–π –∏–∑ 
    - –±–æ–∫—Å–ø–ª–æ—Ç—Ç–∞ –∏ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –¥–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
    - countplot –¥–ª—è –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞.

    –ü–∞—Ä–∞–º–µ—Ç—Ä—ã:
    - df: DataFrame, —Å–æ–¥–µ—Ä–∂–∞—â–∏–π –¥–∞–Ω–Ω—ã–µ –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏.
    - feature: str, –Ω–∞–∑–≤–∞–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–∞ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤.
    - xlabel: str, –º–µ—Ç–∫–∞ –¥–ª—è –æ—Å–∏ X (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é None).
    - bins: int, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ—Ä–∑–∏–Ω –¥–ª—è –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã (–ø–æ —É–º–æ–ª—á–∞–Ω–∏—é 50).
    - hue: str, bool, –¥–æ–±–∞–≤–ª—è–µ—Ç –≤ –≥—Ä–∞—Ñ–∏–∫ —Ä–∞–∑–±–∏–µ–Ω–∏–µ –ø–æ –ø—Ä–∏–∑–Ω–∞–∫—É(str=feature) –∏–ª–∏ –≤—Å–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º –ø—Ä–∏–∑–Ω–∞–∫–∞–º (hue=True), 
    –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –Ω–µ –±–æ–ª–µ–µ 5

    –í–æ–∑–≤—Ä–∞—â–∞–µ–º–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: –û—Ç–æ–±—Ä–∞–∂–∞–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ –∏ –≤—ã–≤–æ–¥–∏—Ç –≥—Ä–∞—Ñ–∏–∫–∏ –Ω–∞ —ç–∫—Ä–∞–Ω.
    """
    
    # –°—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Å–ø–∏—Å–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, 
    # –≥–¥–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –º–µ–Ω—å—à–µ –ª–∏–∏–±–æ —Ä–∞–≤–Ω–æ 5 (—á—Ç–æ–±—ã –Ω–µ –ø–µ—Ä–µ–≥—Ä—É–∂–∞—Ç—å –≥—Ä–∞—Ñ–∏–∫)
    cat_features = [feat for feat in df.columns if df[feat].dtype == 'object' and df[feat].nunique() <=5]

    # –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ
    uniq = df[feature].nunique()
    mean = df[feature].mean()
    median = df[feature].median()
    std = df[feature].std()
    min_val = df[feature].min()
    max_val = df[feature].max()
    q1 = df[feature].quantile(0.25)
    q3 = df[feature].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
    
    print(f"–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è {feature}:")
    print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: {uniq:.0f}")
    print(f"–°—Ä–µ–¥–Ω–µ–µ: {mean:.2f}")
    print(f"–ú–µ–¥–∏–∞–Ω–∞: {median:.2f}")
    print(f"–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {std:.2f}")
    print(f"–ú–∏–Ω–∏–º—É–º: {min_val:.2f}")
    print(f"–ú–∞–∫—Å–∏–º—É–º: {max_val:.2f}")
    
    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
    sns.set(font_scale=1)
    
    # –ì—Ä–∞—Ñ–∏–∫ –¥–ª—è –¥–∏—Å–∫—Ä–µ—Ç–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
    if uniq <= 20:
        if hue == None or isinstance(hue, str):
            plt.figure(figsize = (10, 5))
            sns.countplot(data=df, 
                         x=feature,
                         color='skyblue', 
                         edgecolor='black',
                         hue=hue)
            plt.axvline(x=mean-min_val, color='b', ls='--', lw=2, label='Mean')
            plt.axvline(x=median-min_val, color='r', label='Median')
            plt.legend()
            plt.xticks(rotation=45)
            if title: plt.title(title)
            plt.show()
            
            # –ì—Ä–∞—Ñ–∏–∫–∏ —Å —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ–º –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º –ø—Ä–∏–∑–Ω–∞–∫–∞–º
            fig, axes = plt.subplots(nrows=len(cat_features),
                                     ncols= 1,
                                     figsize=(14, 15))
            axes = axes.flatten()
            
            for index, feat in enumerate(cat_features):
                handles = []
                labels = []
                ax = axes[index]

                hist = sns.countplot(data=df, 
                             x=feature, 
                             color='skyblue', 
                             edgecolor='black',
                             hue=feat,
                             ax=ax)


                for cat in df[feat].unique():
                    mean_value = df[df[feat] == cat][feature].mean()
                    line_handle = ax.axvline(x=mean_value-min_val, 
                                             color=f'#{random.randint(0, 0xFFFFFF):06x}',
                                             ls='--', 
                                             lw=2)
                    handles.append(line_handle)
                    labels.append(f'Mean {feat}={cat}')

                ax.set_title(f'{title} –ø–æ –ø—Ä–∏–∑–Ω–∞–∫—É {feat}')

                ax.legend(handles, labels)

            plt.tight_layout()
            plt.show()
            
    # –ì—Ä–∞—Ñ–∏–∫ –¥–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    else:
        print(f"Q1: {q1:.2f}")
        print(f"Q3: {q3:.2f}")
        print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±—Ä–æ—Å–æ–≤: {len(outliers)}")
        if hue == None or isinstance(hue, str):
            f2, (ax_box2, ax_hist2) = plt.subplots(2, 
                                                   sharex=True, 
                                                   gridspec_kw={"height_ratios": (.15, .85)}, 
                                                   figsize=(12,6))
            sns.boxplot(data=df,
                        x=feature,
                        color='skyblue',
                        ax=ax_box2)
            sns.histplot(data=df, 
                         x=feature,
                         ax=ax_hist2, 
                         color='skyblue', 
                         edgecolor='black',
                         bins=bins,
                         hue=hue,
                         multiple='stack',
                         legend=True,
                         kde=True)
            plt.axvline(x=mean, color='b', ls='--', lw=2, label='Mean')
            plt.axvline(x=median, color='r', label='Median')
            plt.legend()
            plt.xticks(rotation=45)
            if xlabel: ax_hist2.set(xlabel=xlabel)
            if title: ax_box2.set(title=title)
            plt.show()

        elif hue:
            # –û—Å–Ω–æ–≤–Ω–æ–π –≥—Ä–∞—Ñ–∏–∫
            f2, (ax_box2, ax_hist2) = plt.subplots(2, 
                                                   sharex=True, 
                                                   gridspec_kw={"height_ratios": (.15, .85)}, 
                                                   figsize=(14, 6))

            sns.boxplot(data=df,
                    x=feature,
                    color='skyblue',
                    ax=ax_box2)
            viz_box_plot = False
            sns.histplot(data=df, 
                     x=feature,
                     ax=ax_hist2, 
                     color='skyblue', 
                     edgecolor='black',
                     bins=bins,
                     multiple='stack',
                     legend=True,
                     kde=True)
            plt.axvline(x=mean, color='b', ls='--', lw=2, label='Mean')
            plt.axvline(x=median, color='r', label='Median')
            plt.legend()
            plt.xticks(rotation=45)
            if xlabel: ax_hist2.set(xlabel=xlabel)
            if title: ax_box2.set(title=title)

            # –ì—Ä–∞—Ñ–∏–∫–∏ —Å —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ–º –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º –ø—Ä–∏–∑–Ω–∞–∫–∞–º
            fig, axes = plt.subplots(nrows=len(cat_features),
                                     ncols= 1,
                                     figsize=(12, 15))
            axes = axes.flatten()

            for index, feat in enumerate(cat_features):
                handles = []
                labels = []
                ax = axes[index]

                hist = sns.histplot(data=df, 
                             x=feature, 
                             color='skyblue', 
                             edgecolor='black',
                             bins=bins,
                             hue=feat,
                             multiple='stack',
                             legend=True,
                             ax=ax)


                for cat in df[feat].unique():
                    mean_value = df[df[feat] == cat][feature].mean()
                    line_handle = ax.axvline(x=mean_value, 
                                             color=f'#{random.randint(0, 0xFFFFFF):06x}',
                                             ls='--', 
                                             lw=2)
                    handles.append(line_handle)
                    labels.append(f'Mean {feat}={cat}')

                ax.set_title(f'{title} –ø–æ –ø—Ä–∏–∑–Ω–∞–∫—É {feat}')

                ax.legend(handles, labels)

            plt.tight_layout()
            plt.show()
```

## –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü

–í—ã–ø–æ–ª–Ω–∏–º –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü "market_df" —Å —Ç–∞–±–ª–∏—Ü–∞–º–∏ "market_money_df", "market_time_df" –∏ "money_df" –ø–æ "id" –ø–æ –æ—á–µ—Ä–µ–¥–∏.

```python
# –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Ç–∞–±–ª–∏—Ü—É "market_time_df" –≤—ã–¥–µ–ª–µ–Ω–∏–µ–º –ø–µ—Ä–∏–æ–¥–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü—ã
market_time_df_pivot = market_time_df.pivot_table(index='id', 
                                                  columns='–ø–µ—Ä–∏–æ–¥', 
                                                  values='–º–∏–Ω—É—Ç', 
                                                  aggfunc='sum').add_prefix('–º–∏–Ω—É—Ç_')
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã = {market_time_df_pivot.shape[0]}")
market_time_df_pivot.head()
```

```python
# –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Ç–∞–±–ª–∏—Ü—É "market_money_df" –≤—ã–¥–µ–ª–µ–Ω–∏–µ–º –ø–µ—Ä–∏–æ–¥–æ–≤ –≤ —Å—Ç–æ–ª–±—Ü—ã
market_money_df_pivot = market_money_df.pivot_table(index='id', 
                                                    columns='–ø–µ—Ä–∏–æ–¥', 
                                                    values='–≤—ã—Ä—É—á–∫–∞', 
                                                    aggfunc='sum').add_prefix('–≤—ã—Ä—É—á–∫–∞_')

print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ —Å–≥—Ä—É–ø–ø–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —Ç–∞–±–ª–∏—Ü—ã = {market_money_df_pivot.shape[0]}")
market_money_df_pivot.head()
```

```python
# –í—ã–ø–æ–ª–Ω–∏–º –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ "market_df" –∏ "market_time_df_pivot"
market_df_joined = pd.merge(market_df, market_time_df_pivot, on ='id', how='left')

print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –¥–æ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è = {market_df.shape[0]}")
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –ø–æ—Å–ª–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è = {market_df_joined.shape[0]}")
market_df_joined.head()
```

```python
# –í—ã–ø–æ–ª–Ω–∏–º –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ "market_df" –∏ "market_money_df_pivot"
print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –¥–æ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è = {market_df_joined.shape[0]}")

market_df_joined = pd.merge(market_df_joined, market_money_df_pivot, on ='id', how='left')

print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫ –ø–æ—Å–ª–µ –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏—è = {market_df_joined.shape[0]}")
market_df_joined.head()
```

–í–æ-–≤—Ç–æ—Ä—ã—Ö, –ø—Ä–∏ –∏—Ö –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ –±—É–¥–µ—Ç –Ω–∞–±–ª—é–¥–∞—Ç—å—Å—è —Å–ª–µ–¥—É—é—â–∞—è –ø—Ä–æ–±–ª–µ–º–∞:  –∏ –ø—Ä–æ–ø—É—â–µ–Ω–Ω–∞—è (dropped) –ø–µ—Ä–≤–∞—è –∫–∞—Ç–µ–≥–æ—Ä–∏—è, –∏ –ª—é–±—ã–µ –Ω–µ–∏–∑–≤–µ—Å—Ç–Ω—ã–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –±—É–¥—É—Ç –∫–æ–¥–∏—Ä–æ–≤–∞—Ç—å—Å—è –æ–¥–∏–Ω–∞–∫–æ–≤–æ - –≤–µ–∫—Ç–æ—Ä–æ–º –∏–∑ –≤—Å–µ—Ö –Ω—É–ª–µ–π. –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ —Å—Ç–∞–Ω–µ—Ç –Ω–µ–≤–æ–∑–º–æ–∂–Ω—ã–º —Ä–∞–∑–ª–∏—á–∏—Ç—å –∏—Ö –ø–æ—Å–ª–µ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∞—Ü–∏–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

```python
# –ü–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –∏–∑ —Å–ø–∏—Å–∫–∞ ord_columns
# - –∑–∞–ø–æ–ª–Ω–µ–Ω–∏–µ –ø—Ä–æ–ø—É—Å–∫–æ–≤
# - Ordinal-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ

ord_pipe = Pipeline(
    [
        ('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan,
                                                  strategy='most_frequent')),
        ('ord', OrdinalEncoder(categories=[['—Å—Ç–∞–Ω–¥–∞—Ä—Ç', '–ø—Ä–µ–º–∏—É–º'], 
                                           ['–Ω–∏–∑–∫–∏–π', '—Å—Ä–µ–¥–Ω–∏–π', '–≤—ã—Å–æ–∫–∏–π']
                                          ],
                               handle_unknown='use_encoded_value',
                               unknown_value=np.nan)),
        ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)
```

```python
num_pipe = Pipeline(
    [
    ('scaler', StandardScaler()),
    ('poly', PolynomialFeatures(2))
    ]
)
```

```python
# –ü–∞–π–ø–ª–∞–π–Ω –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', num_pipe, num_columns)
    ],
    remainder='passthrough'
)
```

–û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä–æ–≤ –∏ –ø–∞–π–ø–ª–∞–π–Ω–æ–≤ –ø–æ–∑–≤–æ–ª—è–µ—Ç —Å–æ–∑–¥–∞–≤–∞—Ç—å —á–∏—Å—Ç—ã–µ, —Å—Ç—Ä—É–∫—Ç—É—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∏ –≥–∏–±–∫–∏–µ —Ä–∞–±–æ—á–∏–µ –ø—Ä–æ—Ü–µ—Å—Å—ã –¥–ª—è –ø–æ–¥–≥–æ—Ç–æ–≤–∫–∏ –¥–∞–Ω–Ω—ã—Ö. –ù–∞–ø—Ä–∏–º–µ—Ä, —Ç—ã –º–æ–∂–µ—à—å —Å–æ–∑–¥–∞—Ç—å —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä—ã –¥–ª—è –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏ —á–∏—Å–ª–æ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π, –∞ –∑–∞—Ç–µ–º –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –ø–∞–π–ø–ª–∞–π–Ω, –∫–æ—Ç–æ—Ä—ã–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –ø—Ä–∏–º–µ–Ω—è–µ—Ç —ç—Ç–∏ —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä—ã –∏ –∑–∞–≤–µ—Ä—à–∞–µ—Ç—Å—è –æ–±—É—á–µ–Ω–∏–µ–º –º–æ–¥–µ–ª–∏. –≠—Ç–æ —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ —É–ø—Ä–æ—Å—Ç–∏—Ç —É–ø—Ä–∞–≤–ª–µ–Ω–∏–µ –∏ –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö –ø–µ—Ä–µ–¥ –æ–±—É—á–µ–Ω–∏–µ–º –º–æ–¥–µ–ª–µ–π –º–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è.

–ù–µ—Å–∫–æ–ª—å–∫–æ —Å—Å—ã–ª–æ–∫ (*–Ω—É–∂–µ–Ω vpn –¥–ª—è –ø–æ–¥–∫–ª—é—á–µ–Ω–∏—è –∫ —Å–∞–π—Ç–∞–º + –≤–æ–∑–º–æ–∂–Ω–æ –ø–æ–Ω–∞–¥–æ–±–∏—Ç—å—Å—è –ø–æ–∏—Å–∫–æ–º –ø–æ–∏—Å–∫–∞—Ç—å –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–∞—Ç–µ–π - –µ—Å–ª–∏ –æ–Ω–∏ –±—ã–ª–∏ –ø–µ—Ä–µ–º–µ—â–µ–Ω—ã*), –∫–æ—Ç–æ—Ä—ã–µ –ø–æ–∑–≤–æ–ª—è—Ç –ø–æ–≥—Ä—É–∑–∏—Ç—å—Å—è –≤ —Ç–µ–º—É:

- –±–æ–ª—å—à–æ–π –≥–∞–π–¥ –∫–∞–∫ –º–æ–∂–Ω–æ –∫–æ–º–±–∏–Ω–∏—Ä–æ–≤–∞—Ç—å —Ä–∞–∑–Ω—ã–µ —ç—Ç–∞–ø—ã –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ [Pipelines and composite estimators](https://scikit-learn.org/stable/modules/compose.html)
- –∏–ª–∏ –º–æ–∂–Ω–æ –Ω–∞–ø–∏—Å–∞—Ç—å —Å–≤–æ–π –∫–ª–∞—Å—Å-—Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä [Custom Transformers and Pipelines in Python](https://towardsdatascience.com/custom-transformers-in-python-part-ii-6fe111fc82e4)

```python
# –ò—Ç–æ–≥–æ–≤—ã–π –ø–∞–π–ø–ª–∞–π–Ω
final_pipe = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])
```

```python
param_grid = [
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 6),
        'models__max_features': range(2, 6),
        'models__min_samples_split': range(1, 5),
        'models__min_samples_leaf': range(1, 4),
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']  
    },
    
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 6),
        'models__weights': ['uniform', 'distance'],
        'models__algorithm': ['ball_tree', 'kd_tree'],
        'models__p': [1, 2],
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']   
    },

    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear'
        )],
        'models__C': range(1, 5),
        'models__penalty': ['l1', 'l2', 'elasticnet'],
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), RobustScaler(), 'passthrough']  
    },
    
    # —Å–ª–æ–≤–∞—Ä—å –¥–ª—è –º–æ–¥–µ–ª–∏ SVC()
    {
        'models': [SVC(random_state=RANDOM_STATE)],
        'models__kernel': ['sigmoid', 'poly'],
        'preprocessor__num__scaler': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    }
]

grid_search = GridSearchCV(
    final_pipe, 
    param_grid, 
    cv=5,
    scoring='roc_auc',
    n_jobs=-1
)
grid_search.fit(X_train, y_train)
print('–ü–æ–∏—Å–∫ –∑–∞–≤–µ—Ä—à–µ–Ω')
```

```python
# –ü–æ–≤—Ç–æ—Ä–Ω–æ –æ–±—ä–µ–¥–∏–Ω–∏–º –¥–∞—Ç–∞—Å–µ—Ç, –¥–ª—è —Ç–æ–≥–æ —á—Ç–æ–±—ã –≤–µ—Ä–Ω—É—Ç—å –≤—Å–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
market_df_joined_2 = pd.merge(market_df, market_time_df_pivot, on ='id', how='left')
market_df_joined_2 = pd.merge(market_df_joined_2, market_money_df_pivot, on ='id', how='left')
market_df_joined_2 = pd.merge(market_df_joined_2, money_df, on ='id', how='left')
print(market_df_joined_2.shape)
```
